```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

%matplotlib inline
```

```python
# Import iris datasets
df = pd.read_csv('iris.data.txt', header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class'])
df.head()
```

```python
X = df.drop(['class'], axis=1)
```

```python
y = df['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
clf = KNeighborsClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_predict = clf.predict(X_test)
```

```python
pd.DataFrame([y_predict, y_test], ['Prediction Result', 'Original Result']).T.head(20)
```

```python
print('{}%'.format(round(accuracy_score(y_test, y_predict)*100), 5))
```
